# BUILDING THE APP

```python
# main search program



def open_movie_link(event):
    item = result_tree.selection()[0]
    movie_link = result_tree.item(item, "values")[5]
    webbrowser.open_new(movie_link)

def autocomplete(event, entry_widget, listbox_widget):
    value = entry_widget.get().strip().lower()
    all_values = df[entry_widget.entry_name].astype(str).str.lower().unique()
    listbox_widget.delete(0, tk.END)
    for val in all_values:
        if value in val:
            listbox_widget.insert(tk.END, val)

def search_games():
    name = name_var.get().strip().lower()
    genre = genre_var.get()
    min_age = min_age_var.get()
    max_age = max_age_var.get()
    min_price = min_price_var.get()
    max_price = max_price_var.get()
    platform = platform_var.get()

    try:
        min_age = int(min_age)
        max_age = int(max_age)
        min_price = float(min_price)
        max_price = float(max_price)
    except ValueError:
        result_label.config(text="Please enter valid Numerical values for age and price.")
        return
    
    filtered_df = df[
        df['Genres'].str.contains(genre, case=False, na=False) &
        (df['Required age'] >= min_age) &
        (df['Required age'] <= max_age) &
        (df['Price'] >= min_price) &
        (df['Price'] <= max_price) &
        (df[platform] == True)
    ]
    
    if name:
        filtered_df = filtered_df[filtered_df['Name'].str.lower().str.contains(name)]
    
    num_games = len(filtered_df)
    
    if num_games == 0:
        result_label.config(text="No games found")
    else:
        result_label.config(text=f"{num_games} Games found.")
        result_tree.delete(*result_tree.get_children())
        for i, row in filtered_df.iterrows():
            result_tree.insert("", 'end', text=i, values=(
                row["Name"], row["Genres"], row["Release date"],
                row["Required age"], row["Price"], row["Movies"]
            ))

def sort_treeview_column(col):
    data = [(result_tree.set(child, col), child) for child in result_tree.get_children('')]
    data.sort(reverse=ascending[col])
    for index, item in enumerate(data):
        result_tree.move(item[1], '', index)

    ascending[col] = not ascending[col]

def copy_game_name(event=None):
    item = result_tree.selection()[0]
    game_name = result_tree.item(item, "values")[0]
    root.clipboard_clear()
    root.clipboard_append(game_name)

def show_context_menu(event):
    context_menu.tk_popup(event.x_root, event.y_root)

# Main window
root = tk.Tk()
root.title("Game Search")
root.geometry("1250x600")

# Apply a theme
style = ttk.Style(root)
style.theme_use('clam')

# Custom styles
style.configure("TLabel", font=("Helvetica", 12))
style.configure("TEntry", font=("Helvetica", 12))
style.configure("TButton", font=("Helvetica", 12))
style.configure("TOptionMenu", font=("Helvetica", 12))
style.configure("Treeview.Heading", font=("Helvetica", 12, "bold"))
style.configure("Treeview", font=("Helvetica", 12), rowheight=25)

# Entry variables
name_var = tk.StringVar()
genre_var = tk.StringVar()
min_age_var = tk.StringVar()
max_age_var = tk.StringVar()
min_price_var = tk.StringVar()
max_price_var = tk.StringVar()
platform_var = tk.StringVar(value="Windows")

# Search Criteria Frame
search_frame = ttk.Frame(root, padding=(20, 10))
search_frame.grid(row=0, column=0, padx=10, pady=10, sticky='ew')

# Name
name_label = ttk.Label(search_frame, text="Name:")
name_label.grid(row=0, column=0, sticky='w', padx=(0, 10))
name_entry = ttk.Entry(search_frame, textvariable=name_var, width=20)
name_entry.grid(row=0, column=1, sticky='w', padx=(0, 10))

# Genre
genre_label = ttk.Label(search_frame, text="Genre:")
genre_label.grid(row=1, column=0, sticky='w', padx=(0, 10))
genre_entry = ttk.Entry(search_frame, textvariable=genre_var, width=20)
genre_entry.entry_name = "Genres"
genre_entry.grid(row=1, column=1, sticky='w', padx=(0, 10))
genre_listbox = tk.Listbox(search_frame, height=5, width=60)  # Adjust the width as needed
genre_listbox.grid(row=1, column=2, sticky='nsew')
genre_entry.bind('<KeyRelease>', lambda event: autocomplete(event, genre_entry, genre_listbox))

# Age
min_age_label = ttk.Label(search_frame, text="Min Age:")
min_age_label.grid(row=2, column=0, sticky='w', padx=(0, 10))
min_age_entry = ttk.Entry(search_frame, textvariable=min_age_var, width=10)
min_age_entry.grid(row=2, column=1, sticky='w', padx=(0, 10))

max_age_label = ttk.Label(search_frame, text="Max Age:")
max_age_label.grid(row=2, column=2, sticky='w', padx=(0, 10))
max_age_entry = ttk.Entry(search_frame, textvariable=max_age_var, width=10)
max_age_entry.grid(row=2, column=3, sticky='w', padx=(0, 10))

# Price
min_price_label = ttk.Label(search_frame, text="Min Price:")
min_price_label.grid(row=3, column=0, sticky='w', padx=(0, 10))
min_price_entry = ttk.Entry(search_frame, textvariable=min_price_var, width=10)
min_price_entry.grid(row=3, column=1, sticky='w', padx=(0, 10))

max_price_label = ttk.Label(search_frame, text="Max Price:")
max_price_label.grid(row=3, column=2, sticky='w', padx=(0, 10))
max_price_entry = ttk.Entry(search_frame, textvariable=max_price_var, width=10)
max_price_entry.grid(row=3, column=3, sticky='w', padx=(0, 10))

# Platform
platform_label = ttk.Label(search_frame, text="Platform:")
platform_label.grid(row=4, column=0, sticky='w', padx=(0, 10))

platform_frame = ttk.Frame(search_frame)
platform_frame.grid(row=4, column=1, sticky='w', padx=(0, 10))
platform_options = ["Windows", "Mac", "Linux"]
for i, option in enumerate(platform_options):
    rb = ttk.Radiobutton(platform_frame, text=option, variable=platform_var, value=option)
    rb.grid(row=0, column=i, padx=(0, 5))

# Search Button
search_button = ttk.Button(root, text="Search", command=search_games)
search_button.grid(row=1, column=0, padx=10, pady=10)

# Result Frame with Scrollbar
result_frame = ttk.Frame(root)
result_frame.grid(row=2, column=0, padx=10, pady=10, sticky="nsew")

result_tree = ttk.Treeview(result_frame, columns=("Name", "Genres", "Release date", "Age", "Price", "Movies"), show="headings", height=10)
result_tree.heading("Name", text="Name", anchor=tk.CENTER, command=lambda: sort_treeview_column("Name"))
result_tree.heading("Genres", text="Genres", anchor=tk.CENTER, command=lambda: sort_treeview_column("Genres"))
result_tree.heading("Release date", text="Release date", anchor=tk.CENTER, command=lambda: sort_treeview_column("Release date"))
result_tree.heading("Age", text="Age", anchor=tk.CENTER, command=lambda: sort_treeview_column("Age"))
result_tree.heading("Price", text="Price", anchor=tk.CENTER, command=lambda: sort_treeview_column("Price"))
result_tree.heading("Movies", text="Trailer", anchor=tk.CENTER, command=lambda: sort_treeview_column("Movies"))

result_tree.column("Movies", width=200)  # Adjust the width here
result_tree.grid(row=0, column=0, sticky="nsew")

result_scrollbar = ttk.Scrollbar(result_frame, orient="vertical", command=result_tree.yview)
result_scrollbar.grid(row=0, column=1, sticky="ns")
result_tree.configure(yscrollcommand=result_scrollbar.set)

result_label = ttk.Label(root, text="")
result_label.grid(row=3, column=0, padx=10, pady=10)

# Bind double-click event to open movie link
result_tree.bind("<Double-1>", open_movie_link)

# Context menu for copying game name
context_menu = tk.Menu(root, tearoff=0)
context_menu.add_command(label="Copy Game Name", command=copy_game_name)

# Bind right-click to show context menu
result_tree.bind("<Button-3>", show_context_menu)

# Initialize ascending order dictionary for columns
ascending = {
    "Name": True,
    "Genres": True,
    "Release date": True,
    "Age": True,
    "Price": True,
    "Movies": True
}

# Adjust column and row weights
root.columnconfigure(0, weight=1)
root.rowconfigure(2, weight=1)

# Run the main application
root.mainloop()



def visualize_results(game_name, evaluation_result):
    if evaluation_result is not None:
        # Retrieve game data from the DataFrame
        game_data = df[df['Name'] == game_name]
        if not game_data.empty:
            # Create a new window for displaying game information
            info_window = tk.Toplevel()
            info_window.title("Game Information")

            # Set the icon for the window
            info_window.iconbitmap('nod_pic.ico')

            # Add labels for game information
            about_text = tk.Text(info_window, wrap="word", height=10, padx=10, pady=10, font=tkfont.Font(size=12))
            about_text.insert(tk.END, f"About the game: {game_data['About the game'].iloc[0]}")
            about_text.pack(anchor='w', fill='x', padx=10, pady=5)

            info_frame = tk.Frame(info_window)
            info_frame.pack(anchor='w', fill='x', padx=10, pady=5)

            # Function to wrap long text in labels
            def wrap_label_text(label, text):
                if len(text) > 50:  # Adjust this threshold as needed
                    wrapped_text = textwrap.fill(text, width=50)
                    label.config(text=wrapped_text, wraplength=500)  # Adjust wrap length as needed
                else:
                    label.config(text=text)

            # Labels for various game information
            languages_label = tk.Label(info_frame, text="Supported languages:", font=tkfont.Font(size=12))
            languages_label.pack(anchor='w', pady=2)
            languages_text = tk.Label(info_frame, text=game_data['Supported languages'].iloc[0], font=tkfont.Font(size=12))
            languages_text.pack(anchor='w', pady=2)
            wrap_label_text(languages_text, game_data['Supported languages'].iloc[0])

            publishers_label = tk.Label(info_frame, text="Publisher:", font=tkfont.Font(size=12))
            publishers_label.pack(anchor='w', pady=2)
            publishers_text = tk.Label(info_frame, text=game_data['Publishers'].iloc[0], font=tkfont.Font(size=12))
            publishers_text.pack(anchor='w', pady=2)
            wrap_label_text(publishers_text, game_data['Publishers'].iloc[0])

            developers_label = tk.Label(info_frame, text="Developers:", font=tkfont.Font(size=12))
            developers_label.pack(anchor='w', pady=2)
            developers_text = tk.Label(info_frame, text=game_data['Developers'].iloc[0], font=tkfont.Font(size=12))
            developers_text.pack(anchor='w', pady=2)
            wrap_label_text(developers_text, game_data['Developers'].iloc[0])

            tk.Label(info_frame, text=f"Estimated owners: {game_data['Estimated owners'].iloc[0]}", font=tkfont.Font(size=12)).pack(anchor='w', pady=2)
            tk.Label(info_frame, text=f"Categories: {game_data['Categories'].iloc[0]}", font=tkfont.Font(size=12)).pack(anchor='w', pady=2)
            tk.Label(info_frame, text=f"AppID: {game_data['AppID'].iloc[0]}", font=tkfont.Font(size=12)).pack(anchor='w', pady=2)

            # Create a figure with subplots for better visualization
            fig, axes = plt.subplots(1, 2, figsize=(12, 5))

            # Plot 1: Positive vs. Negative Reviews
            sns.barplot(x=['Positive', 'Negative'], y=[game_data['Positive'].iloc[0], game_data['Negative'].iloc[0]], ax=axes[0], palette="muted")
            axes[0].set_title('Positive vs. Negative Reviews')
            axes[0].set_ylabel('Number of Reviews')

            # Plot 2: Price Comparison
            sns.barplot(x=[game_name, 'Mean'], y=[game_data['Price'].iloc[0], df['Price'].mean()], ax=axes[1], palette="muted")
            axes[1].set_title('Price Comparison')
            axes[1].set_ylabel('Price')
            axes[1].set_xlabel('')

            plt.tight_layout()
            plt.show()
        else:
            print(f"Game '{game_name}' not found.")
    else:
        print(f"Game '{game_name}' not found.")

def evaluate_game(name):
    # Dummy evaluation result (Replace this with your actual evaluation logic)
    evaluation_result = 0.8  # Positive-to-Negative Ratio
    # Visualize the evaluation results
    visualize_results(name, evaluation_result)

# Main window
root = tk.Tk()
root.title("Game Evaluation")

# Set the icon for the main window
root.iconbitmap('nod_pic.ico')

# Set custom fonts
label_font = tkfont.Font(size=12, weight='bold')
entry_font = tkfont.Font(size=12)

# Entry for game name
name_label = tk.Label(root, text="Game Name:", font=label_font)
name_label.pack(anchor='w', padx=10, pady=5)
name_entry = tk.Entry(root, font=entry_font)
name_entry.pack(anchor='w', padx=10, pady=5)

# Button to evaluate game
evaluate_button = tk.Button(root, text="Evaluate", command=lambda: evaluate_game(name_entry.get()), font=label_font)
evaluate_button.pack(anchor='w', padx=10, pady=10)

root.mainloop()
```
